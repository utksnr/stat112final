```python
dfmain.columns=(dfmain.columns.str.strip().str.lower().str.replace(" ","_").str.replace("__","_"))
print(dfmain.columns)
#Cleaned column names. Make all letters lower, deleted empty spaces and replaced "__" with "_".
```

```python
dfmain.duplicated().any()
#Check for duplicates.
```

```python
dfmain=dfmain.drop_duplicates()
dfmain.duplicated().any()
#Cleared duplicates.
```

```python
num_cols = ["year", "study_hours_week", "ai_hours_week", "attendance_rate", "assignment_completion_rate", "exam_score", "gpa"]
for i in num_cols:
    if dfmain[i].dtype == 'object':
        dfmain[i] = dfmain[i].str.replace(',', '.')
    
    dfmain[i] = pd.to_numeric(dfmain[i])
#replaced "," with "." in numerical columns.
```

```python
dfmain = dfmain.copy()

# 2. Get the string columns
string_columns = dfmain.select_dtypes(include="object").columns

# 3. Apply the cleaning steps
for col in string_columns:
    dfmain[col] = (
        dfmain[col]
        .str.replace("[^a-zA-Z0-9 / ]", "", regex=True)
        .str.strip()
        .str.lower()
    ) 
```

```python
cat_col = ['gender', 'study_level', 'major', 'ai_usage_frequency', 'primary_ai_use']
for col in cat_col:
    print(f"{col}")
    print(dfmain[col].value_counts())
#checked for misspelling
```

```python
#Correcting gender
dfmain['gender'] = dfmain['gender'].replace(['femae', 'fmale', 'feale'], 'female')
dfmain['gender'] = dfmain['gender'].replace(['nonbinary/oter'], 'nonbinary/other')

#Correcting study_level
dfmain['study_level'] = dfmain['study_level'].replace(['undergrduate', 'undergradute', 'undegraduate', 'unergraduate', 'undergraduae', 'undergradate', 'udergraduate', 'underraduate', 'undrgraduate', 'undergaduate', 'undergrauate'], 'undergraduate')
dfmain['study_level'] = dfmain['study_level'].replace(['mastrs', 'maters'], 'masters')

#Correcting major
dfmain['major'] = dfmain['major'].replace(['busines', 'buiness', 'bsiness'], 'business')
dfmain['major'] = dfmain['major'].replace(['engineerig', 'enginering', 'engineeing', 'engineerng'], 'engineering')
dfmain['major'] = dfmain['major'].replace(['computer sience', 'computer cience', 'computer  science'], 'computer science')
dfmain['major'] = dfmain['major'].replace(['eucation', 'educaton'], 'education')
dfmain['major'] = dfmain['major'].replace(['scial sciences', 'social  sciences'], 'social sciences')
dfmain['major'] = dfmain['major'].replace(['health  sciences', 'health scinces'], 'health sciences')
dfmain['major'] = dfmain['major'].replace(['arts    humanities', 'arts  humanities', 'arts humanities'], 'arts humanities')

#Correcting ai_usage_frequency
dfmain['ai_usage_frequency'] = dfmain['ai_usage_frequency'].replace(['wekly', 'weeky', 'weely'], 'weekly')
dfmain['ai_usage_frequency'] = dfmain['ai_usage_frequency'].replace(['daly', 'daiy'], 'daily')
dfmain['ai_usage_frequency'] = dfmain['ai_usage_frequency'].replace(['rrely', 'rarly'], 'rarely')

#Correcting primary_ai_use
dfmain['primary_ai_use'] = dfmain['primary_ai_use'].replace(['writin support', 'writing suppot', 'wrting support'], 'writing support')
dfmain['primary_ai_use'] = dfmain['primary_ai_use'].replace(['concpt explanation', 'conceptexplanation', 'concept xplanation', 'cocept explanation', 'concept explanatin', 'concept  explanation'], 'concept explanation')
dfmain['primary_ai_use'] = dfmain['primary_ai_use'].replace(['codinghelp', 'coding  help'], 'coding help')
dfmain['primary_ai_use'] = dfmain['primary_ai_use'].replace(['research/reaing', 'reseach/reading', 'research/rading'], 'research/reading')
dfmain['primary_ai_use'] = dfmain['primary_ai_use'].replace(['sudying/flashcards', 'studying/lashcards', 'studying/flashcars'], 'studying/flashcards')
dfmain['primary_ai_use'] = dfmain['primary_ai_use'].replace(['oter', 'othr'], 'other')
```

```python
cat_col = ['gender', 'study_level', 'major', 'ai_usage_frequency', 'primary_ai_use']
for col in cat_col:
    print(f"{col}")
    print(dfmain[col].value_counts())
```

```python
num_cols = ["year", "study_hours_week", "ai_hours_week", "attendance_rate", "assignment_completion_rate", "exam_score", "gpa"]
for i in num_cols:
    q1 = dfmain[i].quantile(0.25)
    q3 = dfmain[i].quantile(0.75)
    iqr = q3 - q1
    
    lb = q1 - 1.5 * iqr
    ub = q3 + 1.5 * iqr
    
    outliers = dfmain[(dfmain[i] < lb) | (dfmain[i] > ub)]
    print(f"Column: {i} outliners: {len(outliers)}")
#Checked numerical columns for outliners using IQR.
```

```python
for i in num_cols:
    q1 = dfmain[i].quantile(0.25)
    q3 = dfmain[i].quantile(0.75)
    iqr = q3 - q1 
    lb = q1 - 1.5 * iqr
    ub = q3 + 1.5 * iqr
    
    col_mean = dfmain[i].mean()
    dfmain.loc[(dfmain[i] < lb) | (dfmain[i] > ub), i] = col_mean
#Replaced outliners with mean in numerical columns.
```

```python
for i in dfmain.columns:
    if dfmain[i].dtype == 'object':
        dfmain[i] = dfmain[i].fillna(dfmain[i].mode()[0])
    else:
        dfmain[i] = dfmain[i].fillna(dfmain[i].median())
#Filling missing values. Used median for numerical data and mode for categorical data.
```
