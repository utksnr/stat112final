This notebook presents a Python-based reproduction of the visualization used in the STAT112 final project.
The aim of this analysis is to examine whether average exam scores differ across AI usage frequency groups.
The dataset used in this notebook is the cleaned version of the original survey data

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Data Loading

In this step, the cleaned dataset is loaded into the notebook using pandas.
The dataset contains information on students' AI usage frequency and exam scores.


```python
df = pd.read_csv("Cleaned_ai_usage_112.csv", sep=';')
df.head()
```

### Computing Group Averages

To address the research question, the average exam score is calculated for each AI usage frequency group.
This allows for a comparison of mean exam performance across different levels of AI usage.


```python
avg_scores = df.groupby("ai_usage_frequency")["exam_score"].mean()
avg_scores

order = ["never", "rarely", "weekly", "daily"]
avg_scores = avg_scores.reindex(order)
```

### Visualization

A bar chart is used to visualize the average exam scores across AI usage frequency groups.
To improve visual clarity and make small differences more noticeable, the y-axis is restricted
to a range between 60 and 90. This adjustment is made only for presentation purposes and does not
affect the underlying data.


```python
plt.figure(figsize=(6,4))
avg_scores.plot(kind="bar")
plt.xlabel("AI Usage Frequency")
plt.ylabel("Average Exam Score")
plt.title("Average Exam Score by AI Usage Frequency")
plt.ylim(60, 90)
plt.tight_layout()
plt.show()
```

### Interpretation

The bar chart shows that average exam scores across AI usage frequency groups are very close to one another.
Although minor differences can be observed, the overall variation between groups is limited.
This suggests that AI usage frequency alone is not strongly associated with exam performance.
