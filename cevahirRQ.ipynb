```python
import pandas as pd

df = pd.read_csv("Cleaned_ai_usage_112.csv", sep=";")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))

sns.boxplot(
    data=df,
    x="ai_usage_frequency",
    y="assignment_completion_rate"
)

plt.title("Assignment Completion Rate by AI Usage Frequency")
plt.xlabel("AI Usage Frequency")
plt.ylabel("Assignment Completion Rate")
plt.tight_layout()
plt.show()
```

```python
order = ["never", "rarely", "weekly", "daily"]

plt.figure(figsize=(8,6))

sns.boxplot(
    data=df,
    x="ai_usage_frequency",
    y="assignment_completion_rate",
    order=order
)

plt.title("Assignment Completion Rate by AI Usage Frequency")
plt.xlabel("AI Usage Frequency")
plt.ylabel("Assignment Completion Rate")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

order = ["never", "rarely", "weekly", "daily"]

plt.figure(figsize=(8,6))

sns.barplot(
    data=df,
    x="ai_usage_frequency",
    y="assignment_completion_rate",
    order=order,
    errorbar="sd"   # standart sapma
)

plt.title("Average Assignment Completion Rate by AI Usage Frequency")
plt.xlabel("AI Usage Frequency")
plt.ylabel("Average Assignment Completion Rate")
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the visual style to a clean, academic whitegrid
sns.set_theme(style="whitegrid")

# Initialize the figure size
plt.figure(figsize=(10, 6))

# Define the specific order for the x-axis categories
order = ['never', 'rarely', 'weekly', 'daily']

# Create the boxplot with a single professional color (monochrome for academic publication)
# 'showmeans=True' adds a marker for the mean value, distinct from the median line
ax = sns.boxplot(
    data=df,
    x="ai_usage_frequency",
    y="assignment_completion_rate",
    order=order,
    color="#4c72b0",  # Professional deep blue
    width=0.5,        # Slightly thinner boxes for a cleaner look
    showmeans=True,
    meanprops={
        "marker": "D", 
        "markerfacecolor": "white", 
        "markeredgecolor": "black", 
        "markersize": 6
    }
)

# Add English titles and labels
plt.title("Assignment Completion Rates by AI Usage Frequency", fontsize=14, pad=15)
plt.xlabel("AI Usage Frequency", fontsize=12)
plt.ylabel("Completion Rate (%)", fontsize=12)

# Set y-axis limits to frame the data properly
plt.ylim(0, 105)

# Adjust layout to prevent clipping and display the plot
plt.tight_layout()
plt.show()
```

```python
plt.tight_layout()

plt.savefig("assignment_completion_by_ai_frequency.png", dpi=300, bbox_inches="tight")

plt.show()
```
